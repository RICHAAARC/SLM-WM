# Content-survival observation

Select an A100 runtime, upload the approved non-secret run request to `/content/slm_wm_content_survival_request.json`, configure the fixed Colab Secrets, and execute the cells in order. The last cell is independent and can package a success, preflight failure, OOM, or scientific failure from disk.

```python
import pathlib
import subprocess
import sys
bootstrap_root = pathlib.Path("/content/slm_wm_content_survival_bootstrap")
subprocess.run(["git", "clone", "--branch", "main", "--single-branch", "https://github.com/RICHAAARC/SLM-WM.git", str(bootstrap_root)], check=True)
controller = bootstrap_root / "scripts/run_content_survival_observation_colab.py"
subprocess.run([sys.executable, "-I", str(controller), "bootstrap-public", "--run-request-json", "/content/slm_wm_content_survival_request.json", "--bootstrap-root", str(bootstrap_root)], check=True)
```

```python
subprocess.run([sys.executable, "-I", str(controller), "preflight", "--run-request-json", "/content/slm_wm_content_survival_request.json"], check=True)
```

```python
subprocess.run([sys.executable, "-I", str(controller), "run", "--run-request-json", "/content/slm_wm_content_survival_request.json"], check=True)
```
